## Шаг 3: анализ причин, по которым заказы не доставляются

Определим, сколько заказов в месяц в среднем не доставляется по разным причинам и выведем детализацию данных причин

Во-первых, посмотрим на заказы в статусе __unavailable__ (недоступны) для выявления причин их недоступности

```python
unavailable_orders = orders.query('order_status == "unavailable"')
unavailable_orders.isnull().sum()
```

Здесь можно заметить, что заказы в статусе __unavailable__ не доставлены покупателям из-за того, что после подтверждения оплаты заказы __не переданы в логистическую службу продавцом__. Будем считать это одной из причин того, почему заказ не доставлен покупателю.

Теперь посмотрим на заказы в статусе __canceled__. Возможно, в данных будут пропуски, по которым мы сможем выявить стадии, на которых заказы отменяются.

```python
canceled_orders = orders.query('order_status == "canceled"')
canceled_orders.isnull().sum()
```

Выведем в датафрейм заказы, о которых нет данных

```python
filtered_orders = canceled_orders[canceled_orders.isna().any(axis=1)]
filtered_orders.head()
```

Как и предполагалось ранее, пропущенные значения в датах дают информацию о причинах отмены заказов. И также можно сделать вывод, что при отсутствии даты на каком-либо из этапов, дальнейшие этапы также не производятся. А значит, заказ входит в категорию недоставленных. 

По данной таблице видно, что заказы отменяются на этапах оплаты (__заказ не оплачен покупателем__), передачи в логистический центр (__заказ не передан в логистический центр продавцом__) и доставки (__заказ не доставлен покупателю__). 

Итак, выявлен ряд причин, по которым заказы не доставляются покупателям: 
1. Заказ не оплачен покупателем (статус canceled)
2. Заказ не передан в логистический центр продавцом (статус unavailable)
3. Заказ не передан в логистический центр продавцом из-за отмены заказа (статус canceled)
4. Заказ не доставлен покупателю логистической службой из-за отмены заказа (статус canceled)

Добавим столбец с причиной недоставки заказа и сведем всё в один датафрейм для удобства визуализации данных.

```python
unavailable_orders = unavailable_orders[unavailable_orders.isna().any(axis=1)]
unavailable_orders['reason'] = 'Заказ не передан в логистический центр продавцом и недоступен'
```

Напишем функцию для определения причины недоставки в заказах со статусом canceled

```python
def reason(approved_date, delivered_carrier_date, delivered_customer_date):
    if approved_date == 0:
        return 'Заказ не оплачен покупателем и отменен'
    elif delivered_carrier_date== 0:
        return 'Заказ не передан в логистическую службу продавцом и отменен'
    elif delivered_customer_date== 0:
        return 'Заказ не доставлен логистической службой покупателю и отменен'
```

```python
filtered_orders = filtered_orders.fillna(0)
filtered_orders['reason'] = filtered_orders.apply(lambda x: reason(x['order_approved_at'], x['order_delivered_carrier_date'], x['order_delivered_customer_date']), axis=1)
```

```python
general_reasons = pd.concat([unavailable_orders, filtered_orders], axis=0)
general_reasons = general_reasons.fillna(0)
general_reasons
```

```python
not_delivered_reasons = general_reasons.groupby('reason', as_index=False) \
    .agg({'order_id':'count'}) \
    .rename(columns={'order_id':'count_of_reasons'}) \
    .sort_values('reason', ascending=True)
pd.set_option('max_colwidth', 150)
pd.set_option('display.width', 500)
not_delivered_reasons.head()
```

```python
plt.pie(not_delivered_reasons.count_of_reasons, labels = not_delivered_reasons.reason, 
        wedgeprops={"linewidth": 2, "edgecolor": "white"}, autopct=lambda p : '{:.2f}%  ({:,.0f})'.format(p,p * sum(not_delivered_reasons.count_of_reasons)/100))
plt.show()
```

```python
not_delivered_reasons.count_of_reasons.sum()
```

Таким образом, __за всё время__ работы было __недоставлено 1228 заказов__ по различным причинам. 

Теперь посмотрим на __распределение по месяцам__ и __среднемесячное количество__ _недоставленных_ заказов. Распределение будем смотреть по месяцу, в котором стоит обещанная дата доставки.

```python
def not_delivered_date(approved_date, delivered_carrier_date, delivered_customer_date, estimated_delivery_date):
    if approved_date != 0:
        return pd.to_datetime(approved_date) + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin()
    elif delivered_carrier_date != 0:
        return pd.to_datetime(delivered_carrier_date) + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin()
    elif delivered_customer_date != 0:
        return pd.to_datetime(delivered_customer_date) + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin()
    else:
        return pd.to_datetime(estimated_delivery_date) + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin()

general_reasons['not_delivered_date'] = general_reasons.apply(lambda x: not_delivered_date(x['order_approved_at'], x['order_delivered_carrier_date'], 
                                                                                           x['order_delivered_customer_date'], 
                                                                                           x['order_estimated_delivery_date']), axis=1)
general_reasons.not_delivered_date = pd.to_datetime(general_reasons.not_delivered_date).dt.date

monthly_general_reasons = general_reasons.groupby(['not_delivered_date','reason'], as_index=False) \
    .agg({'order_id':'count'}) \
    .rename(columns={'order_id':'summary_count'})
monthly_general_reasons = monthly_general_reasons.groupby('reason', as_index=False) \
    .agg({'summary_count':'mean'}) \
    .rename(columns={'summary_count':'montly_mean'}) \
    .sort_values('reason', ascending=True)
monthly_general_reasons.montly_mean = round(monthly_general_reasons.montly_mean, 0)
```

```python
general_reasons = general_reasons.groupby(['not_delivered_date', 'reason'], as_index=False) \
    .agg({'order_id':'count'}) \
    .rename(columns={'order_id':'count_orders'}) \
    .sort_values('not_delivered_date', ascending=True)
plt.figure(figsize=(15,8))
ax = sns.lineplot(data=general_reasons, x='not_delivered_date', y='count_orders', hue='reason')
```

__Среднемесячное количество недоставленных заказов по различным причинам__

```python
pd.set_option('max_colwidth', 150)
pd.set_option('display.width', 500)
monthly_general_reasons
```

```python
plt.pie(monthly_general_reasons.montly_mean, labels = monthly_general_reasons.reason, 
        wedgeprops={"linewidth": 2, "edgecolor": "white"})
plt.show()
```

Таким образом, _среднемесячное количество_ заказов, которые _не доставлены_ из-за того, что не были переданы в логистическую службу продавцом и имеют статус 'unavailabel', является наибольшим (в среднем 29 заказов в месяц). При этом, _среднемесячное количество_ недоставленных заказов из-за неуплаты покупателем является наименьшим (в среднем 6 заказов в месяц). 

## Шаг 4: анализ продаж товаров в различных временных срезах

Для каждого вида товара определим самый удачный день недели - __день недели, в который этот вид товара продается чаще всего__.

Днем продажи будет считаться день подтверждения оплаты заказа, покупкой будет считаться заказ, у которого подтверждена оплата (отмечено время подтверждения заказа) и который доставлен пользователю (указано время доставки). 

Данные по заказам (customer_id), которые считаются покупками, хранятся в списке __success_orders__. 

Для опеределения, в какой день товар продается чаще всего, будем использовать _среднее_ по дням недели.

```python
daily_orders = orders.merge(orders_item, on='order_id')
daily_orders = daily_orders.query('customer_id in @success_orders').groupby(['product_id', 'order_approved_at'], as_index=False) \
    .agg({'order_status':'count'}) \
    .rename(columns={'order_status':'count_of_sales'})    
daily_orders.order_approved_at = daily_orders.order_approved_at.dt.day_name()
daily_orders = daily_orders.groupby(['product_id', 'order_approved_at'], as_index=False) \
    .agg({'count_of_sales':'mean'}) \
    .rename(columns={'order_approved_at':'day_of_week'}) \
    .sort_values('count_of_sales', ascending=False) \
    .drop_duplicates(subset='product_id', keep='first')
daily_orders
```

Судя по полученным данным, самые часто продаваемые в определенный день недели товары продаются в четверг, пятницу и понедельник. 

Посмотрим на общую картину продаж по дням недели в среднем.

```python
week_sales = daily_orders.groupby('day_of_week', as_index=False) \
    .agg({'count_of_sales':'mean'}).sort_values('count_of_sales', ascending=False)
week_sales
```

Действительно, большинство продаж происходит в четверг, однако пятница самый непродающий день. 

## Шаг 5: анализ среднего количества покупок пользователей в неделю (по месяцам)

Для определения среднего количества покупок пользователей будем использовать те заказы, которые считаются покупками (customer_id хранится в success_orders). Датой покупки будем считать дату подтверждения оплаты.

```python
weekly_sales = orders.merge(customers, on='customer_id')
weekly_sales = weekly_sales.query('customer_id in @success_orders') \
    .groupby(['customer_unique_id', 'order_approved_at'], as_index=False) \
    .agg({'order_id':'count'}) \
    .rename(columns={'order_id':'count_orders', 'order_approved_at':'date_of_sale'}) \
    .sort_values('date_of_sale', ascending=True)
weekly_sales.date_of_sale = weekly_sales.date_of_sale.dt.date
weekly_sales.date_of_sale = pd.to_datetime(weekly_sales.date_of_sale) + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin()
weekly_sales = weekly_sales.groupby(['customer_unique_id', 'date_of_sale'], as_index=False) \
    .agg({'count_orders':'sum'}) \
    .sort_values('date_of_sale', ascending=True)
weekly_sales['purchases_in_week'] = weekly_sales.apply(lambda x: round(x.count_orders / (((monthrange(x.date_of_sale.year, x.date_of_sale.month))[1])/7),5), axis=1)
weekly_sales
```

```python
round(weekly_sales.purchases_in_week.mean(),3)
```

Можно заметить, что пользователи делают не так много заказов каждую неделю. __Среднее количество покупок в неделю__ по всем пользователям за всё время - __0.233.__ 

## Шаг 5: когортный анализ пользователей 

Проведем когортный анализ пользователей __с января по декабрь__. Судя по полученным данным, покупки совершались в период октября 2016 года до октября 2018 года. Поэтому берем период январь 2017 - декабрь 2017 (полный 2017 год) 

Поделим всех пользователей на когорты по __месяцу их первой покупки__. Месяцом покупки будет считать месяц оплаты заказа.

```python
df_for_cohorts = orders.merge(customers[['customer_id', 'customer_unique_id']], on='customer_id')
df_for_cohorts = df_for_cohorts.query('order_status != "canceled"').sort_values('order_approved_at', ascending=True)
df_for_cohorts = df_for_cohorts[['customer_unique_id', 'order_approved_at']]
df_for_cohorts = df_for_cohorts.dropna()
```

```python
df_for_cohorts.head()
```

```python
df_for_cohorts.dtypes
```

```python
df_for_cohorts = df_for_cohorts[(df_for_cohorts['order_approved_at'] > '31.12.2016') & (df_for_cohorts['order_approved_at'] <= '31.12.2017')]
```

```python
df_for_cohorts
```

```python
first_order = df_for_cohorts.groupby('customer_unique_id', as_index=False) \
    .agg({'order_approved_at':'min'}) \
    .rename(columns={'order_approved_at':'first_order'})
```

```python
first_order.first_order = first_order.first_order.dt.month
```

```python
first_order
```

```python
cohorts_2017 = df_for_cohorts.merge(first_order, on='customer_unique_id')
```

```python
cohorts_2017.order_approved_at = cohorts_2017.order_approved_at.dt.month
```

```python
cohorts_2017['period'] = cohorts_2017['order_approved_at'] - cohorts_2017['first_order']
```

```python
cohorts_2017.head()
```

```python
cohorts = cohorts_2017.pivot_table(index='first_order', columns='period', values='customer_unique_id', aggfunc='count')
cohorts_style = (cohorts
            .style
            .background_gradient(cmap='YlOrRd')
            .highlight_null('white')
            .format('{:.2f}', na_rep=''))
cohorts_style
```

Построили когортный анализ количества пользователей, которые сделали первую покупку в опеределенный месяц 2017 года.
Теперь посмотрим на процентное распределение для определения самого высокого retention

```python
cohort_sizes = cohorts.apply(lambda x: x.dropna().iloc[0], axis=1)
retention_rates = cohorts.div(cohort_sizes, axis=0)
```

```python
cohorts_style = (retention_rates
            .style
            .background_gradient(cmap='YlOrRd')
            .highlight_null('white')
            .format('{:.3%}', na_rep=''))
```

```python
cohorts_style
```

Построили когорты до третьего знака после запятой для выявления когорты с самым высоким retention в 3 месяце. 

```python
max_3 = retention_rates[3].idxmax()
max_3
```

Таким образом, самый высокий retention на 3 месяц имеет когорта пользователей, которые сделали первую покупку в __марте 2017 года.__
